# BASELINE 1 : DummyClassifier

Avant de construire notre modèle, on a besoin d’un référentiel minimal :

“Est-ce que mon modèle fait mieux qu’un modèle stupide ?”

Le DummyClassifier prédit toujours la classe majoritaire

```python
# =====================================================
# BASELINE 1 — DUMMY CLASSIFIER
# =====================================================
# Objectif :
# - Créer une référence minimale
# - Le modèle prédit toujours la classe majoritaire
# - Sert de point de comparaison pour les modèles futurs
# =====================================================

from sklearn.dummy import DummyClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Initialisation du modèle
dummy_clf = DummyClassifier(strategy="most_frequent", random_state=42)

# Entraînement
dummy_clf.fit(X_train_final, y_train)

# Prédiction sur le jeu de test
y_pred_dummy = dummy_clf.predict(X_test_final)

# Résultats
print("=== Dummy Classifier ===\n")
print("Matrice de confusion :")
print(confusion_matrix(y_test, y_pred_dummy))

print("\nRapport de classification :")
print(classification_report(y_test, y_pred_dummy))
```

# BASELINE ML : RÉGRESSION LOGISTIQUE

```python
# =====================================
# BASELINE ML : RÉGRESSION LOGISTIQUE
# =====================================

# --- 1. IMPORTS ---
# Modèle de classification binaire interprétable
from sklearn.linear_model import LogisticRegression

# Métriques d'évaluation adaptées à la classification
from sklearn.metrics import (
    confusion_matrix,      # Pour analyser FN / FP
    classification_report, # Precision / Recall / F1
    roc_auc_score,         # Mesure globale de séparation
    roc_curve              # Pour la courbe ROC
)

# Librairie de visualisation
import matplotlib.pyplot as plt


# --- 2. INITIALISATION DU MODÈLE ---
# max_iter=1000 : assure la convergence du modèle
# class_weight='balanced' : donne plus de poids à la classe minoritaire (Dropout)
# random_state : reproductibilité des résultats
log_reg = LogisticRegression(
    max_iter=1000,
    class_weight='balanced',
    random_state=42
)


# --- 3. ENTRAÎNEMENT DU MODÈLE ---
# Le modèle apprend la relation entre les variables explicatives
# (X_train_final) et la cible (y_train)
log_reg.fit(X_train_final, y_train)


# --- 4. PRÉDICTIONS ---
# Prédiction de la classe (0 ou 1)
y_pred = log_reg.predict(X_test_final)

# Prédiction de la probabilité d'appartenir à la classe 1 (Dropout)
# On garde uniquement la colonne correspondant à la classe 1
y_proba = log_reg.predict_proba(X_test_final)[:, 1]


# --- 5. ÉVALUATION DU MODÈLE ---
# Matrice de confusion : FN / FP / TP / TN
cm = confusion_matrix(y_test, y_pred)

# Rapport détaillé des métriques
report = classification_report(y_test, y_pred)

# Aire sous la courbe ROC
auc = roc_auc_score(y_test, y_proba)


# --- 6. AFFICHAGE DES RÉSULTATS ---
print("=== Régression Logistique ===")

print("\nMatrice de confusion :")
print(cm)

print("\nRapport de classification :")
print(report)

print(f"AUC-ROC : {auc:.3f}")
```

```python
# --- COURBE ROC ---
# Calcul des taux de faux positifs et vrais positifs
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

# Création de la figure
plt.figure(figsize=(6,4))

# Tracé de la courbe ROC
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")

# Diagonale = modèle aléatoire
plt.plot([0,1], [0,1], linestyle='--')

# Labels et titre
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Courbe ROC - Régression Logistique")

# Légende
plt.legend()

# Affichage
plt.show()
```

# MLFlow Regression Logistique

ÉTAPE 1 — Importer MLflow

```python
# MLflow permet de suivre les expériences de machine learning
# (paramètres, métriques, modèles, figures)
import mlflow
import mlflow.sklearn
```

ÉTAPE 2 — Définir une expérience MLflow

```python
# Création (ou récupération si elle existe déjà) d'une expérience MLflow
# Toutes les exécutions (runs) seront regroupées sous ce nom
mlflow.set_experiment("Student Dropout Prediction")
```

ÉTAPE 3 — MLflow : Baseline ML (Régression Logistique simple)

```python
# ======================================
# MLflow — Log du modèle LogisticRegression Baseline
# ======================================

from sklearn.metrics import accuracy_score, recall_score, f1_score, roc_auc_score
import mlflow
import mlflow.sklearn

# -----------------------------
# DÉBUT D'UN RUN MLflow
# -----------------------------
# Chaque run correspond à une exécution unique (train + métriques + modèle)
with mlflow.start_run(run_name="LogisticRegression_Baseline"):

    # -----------------------------
    # PARAMÈTRES DU MODÈLE
    # -----------------------------
    # On log les paramètres importants pour traçabilité
    mlflow.log_param("model_type", "LogisticRegression")  # Nom du modèle
    mlflow.log_param("class_weight", "balanced")          # Gestion des classes déséquilibrées
    mlflow.log_param("max_iter", 1000)                    # Nombre maximum d'itérations pour convergence

    # -----------------------------
    # PRÉDICTIONS
    # -----------------------------
    # Prédiction des classes (0 = non-dropout, 1 = dropout)
    y_pred = log_reg.predict(X_test_final)

    # Prédiction des probabilités pour la classe 1 (Dropout)
    y_proba = log_reg.predict_proba(X_test_final)[:, 1]

    # -----------------------------
    # MÉTRIQUES
    # -----------------------------
    # Accuracy globale
    accuracy = accuracy_score(y_test, y_pred)

    # Recall de la classe 1 (dropout) → priorité métier
    recall_1 = recall_score(y_test, y_pred, pos_label=1)

    # F1-score de la classe 1
    f1_1 = f1_score(y_test, y_pred, pos_label=1)

    # AUC-ROC : capacité globale de séparation
    auc = roc_auc_score(y_test, y_proba)

    # -----------------------------
    # LOGGING DES MÉTRIQUES DANS MLflow
    # -----------------------------
    mlflow.log_metric("accuracy", accuracy)
    mlflow.log_metric("recall_dropout", recall_1)
    mlflow.log_metric("f1_dropout", f1_1)
    mlflow.log_metric("auc", auc)

    # -----------------------------
    # SAUVEGARDE DU MODÈLE DANS MLflow
    # -----------------------------
    # 'name' remplace l'ancien 'artifact_path' pour être à jour
    mlflow.sklearn.log_model(log_reg, name="LogisticRegression_Baseline_Model")

    print("✅ Run MLflow terminé et modèle sauvegardé !")
```

# GridSearchCV

```python
# =====================================================
# GRIDSEARCHCV — RÉGRESSION LOGISTIQUE
# =====================================================
# Objectif :
# - Optimiser les hyperparamètres de la régression logistique
# - Se concentrer sur le F1-score pour la classe 1 (Dropout)
# =====================================================

# --- 1. IMPORTS ---
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer, f1_score, classification_report, confusion_matrix, roc_auc_score

# --- 2. INITIALISATION DU MODÈLE ---
log_reg = LogisticRegression(
    max_iter=1000,           # Convergence garantie
    class_weight='balanced', # Gérer le déséquilibre des classes
    random_state=42
)

# --- 3. DÉFINITION DE LA GRILLE D'HYPERPARAMÈTRES ---
param_grid = {
    'C': [0.01, 0.1, 1, 10],       # Régularisation
    'penalty': ['l1', 'l2'],       # Type de pénalité
    'solver': ['liblinear']         # Compatible L1 et L2
}

# --- 4. SCORER PERSONNALISÉ ---
# On se concentre sur le F1-score de la classe 1 (Dropout)
f1_scorer = make_scorer(f1_score, pos_label=1)

# --- 5. INITIALISATION DU GRIDSEARCH ---
grid_search = GridSearchCV(
    estimator=log_reg,     # Modèle à tester
    param_grid=param_grid, # Grille d'hyperparamètres
    scoring=f1_scorer,     # Critère d'évaluation
    cv=5,                  # 5-fold cross-validation
    verbose=1,             # Affiche l'avancement
    n_jobs=-1              # Utilise tous les cœurs CPU pour accélérer 
)

# --- 6. LANCEMENT DE LA RECHERCHE ---
grid_search.fit(X_train_final, y_train)

# --- 7. MEILLEUR MODÈLE ---
print("✅ GridSearch terminé !")
print("Meilleurs paramètres :", grid_search.best_params_)
print("Meilleur F1-score CV :", grid_search.best_score_)

best_model = grid_search.best_estimator_

# --- 8. ÉVALUATION SUR LE JEU DE TEST ---
y_pred_best = best_model.predict(X_test_final)
y_proba_best = best_model.predict_proba(X_test_final)[:, 1]

print("\n=== GridSearchCV Logistic Regression ===")
print("\nMatrice de confusion :")
print(confusion_matrix(y_test, y_pred_best))

print("\nRapport de classification :")
print(classification_report(y_test, y_pred_best))

print(f"AUC-ROC : {roc_auc_score(y_test, y_proba_best):.3f}")
```


# MLFlow GridSearchCV

```python
# ======================================
# MLflow — Log du meilleur modèle GridSearchCV
# ======================================

from sklearn.metrics import accuracy_score, recall_score, f1_score, roc_auc_score
import mlflow
import mlflow.sklearn

# -----------------------------
# DÉBUT D'UN RUN MLflow
# -----------------------------
# Chaque run correspond à une exécution unique (GridSearchCV)
with mlflow.start_run(run_name="LogisticRegression_GridSearchCV"):

    # -----------------------------
    # PARAMÈTRES DU MEILLEUR MODÈLE
    # -----------------------------
    # On récupère les paramètres qui ont donné le meilleur F1-score
    best_params = grid_search.best_params_
    mlflow.log_param("model_type", "LogisticRegression_GridSearchCV")
    mlflow.log_param("best_C", best_params['C'])
    mlflow.log_param("best_penalty", best_params['penalty'])
    mlflow.log_param("best_solver", best_params['solver'])
    mlflow.log_param("class_weight", "balanced")
    mlflow.log_param("max_iter", 1000)

    # -----------------------------
    # PRÉDICTIONS
    # -----------------------------
    # Meilleur modèle issu de GridSearch
    best_model = grid_search.best_estimator_

    # Prédiction des classes sur le test
    y_pred = best_model.predict(X_test_final)

    # Probabilité d'appartenir à la classe 1 (dropout)
    y_proba = best_model.predict_proba(X_test_final)[:, 1]

    # -----------------------------
    # MÉTRIQUES
    # -----------------------------
    accuracy = accuracy_score(y_test, y_pred)
    recall_1 = recall_score(y_test, y_pred, pos_label=1)
    f1_1 = f1_score(y_test, y_pred, pos_label=1)
    auc = roc_auc_score(y_test, y_proba)

    # -----------------------------
    # LOGGING DES MÉTRIQUES DANS MLflow
    # -----------------------------
    mlflow.log_metric("accuracy", accuracy)
    mlflow.log_metric("recall_dropout", recall_1)
    mlflow.log_metric("f1_dropout", f1_1)
    mlflow.log_metric("auc", auc)

    # -----------------------------
    # SAUVEGARDE DU MEILLEUR MODÈLE
    # -----------------------------
    mlflow.sklearn.log_model(best_model, name="LogisticRegression_GridSearchCV_Model")

    print("✅ Run MLflow GridSearchCV terminé et meilleur modèle sauvegardé !")
```
